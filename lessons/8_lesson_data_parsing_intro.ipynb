# Стабилизирующая конструкция для работы try/except/finally

```python
try:
    url = 'https://www.comicstreet.ru/collection/frontpage'
    browser = webdriver.Chrome(path_CDM)
    browser.get(url) # Открываем страницу
    
except Exception as e:
    print('Ошибка')
    print(e)
    
finally:
    time.sleep(3)  # Ждем 10 секунд
    browser.quit()  # закрываем браузер после всех манипуляций
    print('Закрыл')
    
```

# Request + BeautifulSoup + Selenium

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver

from webdriver_manager.chrome import ChromeDriverManager
```

```python
rubric = '5'
i = 1

news_df = pd.DataFrame(columns=['rubric', 'date', 'url', 'link', 'title', 'text', 'autor'])
news_df
```

```python
from tqdm import tqdm
import bs4

for date in tqdm(pd.date_range('2021-10-11', '2021-10-22').format()):
    
    url = 'https://www.kommersant.ru/archive/rubric/' + rubric + '/day/' + date
    
    with requests.Session() as session:
        
        html = session.get(url).text
        
        try:
            soup = bs4.BeautifulSoup(html, 'html.parser')
            links_divs = soup.findAll('div', {'class': 'archive_result__item_text'})
            links = ['https://kommersant.ru' + link.find('a', href=True)['href'] for link in links_divs]
        
        except AttributeError:
            print(url)
            continue

        for link in links:
            
            try:
                if 'https://kommersant.ru/gallery/' in link:
                    continue

                html = session.get(link).text
                soup = bs4.BeautifulSoup(html,'html.parser')

                if soup.find('h1', {'class': 'article_name'}) is not None:
                    title = soup.find('h1', {'class': 'article_name'}).text
                elif soup.find('h2', {'class': 'article_name'}) is not None:
                    title = soup.find('h2', {'class': 'article_name'}).text
                else:
                    print(url, link)
                    continue
                    
                news_body = soup.find('div', {'class': 'article_text_wrapper'})
                news_autor = news_body.find('p', {'class': 'b-article__text document_authors'})
                if news_autor is not None:
                    autor = news_body.find('p', {'class': 'b-article__text document_authors'}).text
                    news_autor.decompose()
                else:
                    autor = 'NaN'

                divs_list = news_body.findAll('div')
                for div in divs_list:
                    div.decompose()

                text = news_body.text.replace('\n',' ').strip()
                try:
                    text_chech_encode = text.encode()
                    
                except UnicodeEncodeError:
                    print(url, link)
                    continue
                
                tmp_df = pd.DataFrame([[rubric, date, url, link, title, text, autor]], 
                                      columns=['rubric', 'date', 'url', 'link', 'title', 'text', 'autor'])
                news_df = news_df.append(tmp_df)
        
            except AttributeError:
                print(url, link)
                continue
     
    i += 1
    if not i % 30:
        print(date)
        print('Save temporary data...')
        news_df.to_csv('../data/news_rubric_' + rubric + '_' + date + '.csv', index=False)
        
news_df.to_csv('../data/news_rubric_'+ rubric + '_' + '2011-01-01' + '_' + date + '.csv', index=False)
```

Обработка случая "показать еще"

```python
from selenium import webdriver

tmp = news_df.groupby(['url'])['link'].count().reset_index()
urls_30 = tmp[tmp.link == 30].url.tolist()
```

```python
urls_30
```

```python
urls_30_links_df = pd.DataFrame(columns=['url', 'link'])

driver = webdriver.Chrome('../data/chromedriver')

for url in tqdm(urls_30):
    driver.get(url)
    try:
        driver.find_element_by_class_name('lazyload-button').click()
    except:
        continue

    while True: 
        try:
            driver.implicitly_wait(3) 
            driver.find_element_by_class_name('lazyload-button').click()
        except:
            break

    html = driver.page_source
    soup = bs4.BeautifulSoup(html, 'html.parser')
    links_divs = soup.findAll('div', {'class': 'archive_result__item_text'})
    links = ['https://kommersant.ru' + link.find('a', href=True)['href'] for link in links_divs]
    
    tmp = pd.DataFrame([url for _ in range(len(links))], columns=['url'])
    tmp['link'] = links
    urls_30_links_df = urls_30_links_df.append(tmp)
    
driver.close()
urls_30_links_df.to_csv('../data/urls_30_links_df.csv')
```



<div class="alert alert-info">
<b>Класс spider: </b>
    
В Scrapy, Spider является классом, где определяется поведение при анализе (какие ссылки или URL должны пройти через скрапинг) и при скрапинге (что нужно извлечь).
</div>

```python
import scrapy


class BrickSetSpider(scrapy.Spider):
    name = "brickset_spider"
    start_urls = ['https://www.comicstreet.ru/collection/frontpage']
                  
                  
    def parse(self, response):
        SET_SELECTOR = '.set'
        for brickset in response.css(SET_SELECTOR):

            NAME_SELECTOR = 'div ::text'
            yield {
                'titles': brickset.css(NAME_SELECTOR).extract_first(),
            }
```

```python
pars = BrickSetSpider()
pars.parse()
```
